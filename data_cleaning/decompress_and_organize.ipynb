```
# file names

# .json.xz
test_files = ["2020-03-06_21-10-56_UTC.json.xz", 
              "2020-03-06_21-10-46_UTC.json.xz", 
              "2020-03-06_21-10-38_UTC.json.xz"]

# .txt
test_txt_files = ["2020-03-06_21-10-56_UTC.txt",
                  "2020-03-06_21-10-46_UTC.txt",
                  "2020-03-06_21-10-38_UTC.txt"
                  ]

## TODO: other files to include - if they exist for a post
# comments.json

# location.txt

# .jpg (hold off on for now)
```

```
# a list of dictionaries
jsons = []

for file in test_files:
  read = lzma.open(file).read().decode('utf-8')
  jsondict = json.loads(read)
  jsons.append(jsondict["node"])  
```

```
# a list of the post dsecrips as strings
texts = [open(file).read() for file in test_txt_files]
# print(len(texts))
# print(texts[0])

# extract hashtags
# returns a list of lists (of ea post's hashtags)
hashtags = [re.findall(r"#(\w+)", descrip) for descrip in texts]
```

```
txt_fnames = [f.split('.txt')[0] for f in test_txt_files]
jsonxz_fnames = [f.split('.json.xz')[0] for f in test_files]
```

```
textsS = pd.Series(texts)
hashtagsS = pd.Series(hashtags)
txt_fnamesS = pd.Series(txt_fnames)
df_txt = pd.DataFrame(pd.concat([txt_fnamesS,textsS,hashtagsS],axis=1))
df_txt.columns = ['file_name', 'descrip', 'hashtags']
```
